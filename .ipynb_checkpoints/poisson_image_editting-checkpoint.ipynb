# Composition of a region in a source image to a region in the target image


```python
import numpy as np

def build_A(pts1, pts2):
    """Constructs the intermediate matrix A used in the computation of an
    homography mapping pts1 to pts2."""

    # Build and return A

    n = pts1.shape[0]

    extended_pts1 = np.ones((n, 3))
    extended_pts1[:,:2] = pts1
    extended_pts1 = np.tile(extended_pts1.repeat(2, axis=0), reps=(1, 3))

    A1 = np.tile(np.eye(2).repeat(3, axis=1), reps=(n, 1))
    A2 = pts2.flatten().reshape(-1, 1).repeat(3, axis=1)

    A = np.concatenate((A1, -A2), axis=1) * extended_pts1

    return A

def compute_H(pts1, pts2):
    """Computes an homography mapping one set of co-planar points (pts1) to another (pts2)."""

    # Construct the intermediate A matrix.
    A = build_A(pts1, pts2)
    # Compute the symmetric matrix AtA.
    AtA = A.T @ A
    # Compute the eigenvalues and eigenvectors of AtA.
    eig_w, eig_v = np.linalg.eigh(AtA)
    # Return the eigenvector corresponding to the smallest eigenvalue
    smallest_eig_v = eig_v[:, np.argmin(eig_w)] # take the i-th column of eig_v for i = argmin of eig_w
    # reshape as a 3x3 matrix.
    H = smallest_eig_v.reshape(3,3)

    return H

# This is a supplementary function
# for computing pts2 given H and pts1
def project_Homo(H, pts1):
  """Re-compute pts2 given H and pts1"""
  n = pts1.shape[0]
  # convert pts1 to homogeneous coordinates
  PTS1 = np.ones((n, 3))
  PTS1[:, :2] = pts1
  PTS1 = PTS1.T

  # projecting
  PTS2 = H @ PTS1
  PTS2 = PTS2.T
  norm = PTS2[:, 2].reshape(-1, 1)
  # convert back to homogeneous coordinates
  PTS2 = PTS2 / norm

  return PTS2[:, :2]

def bilinear_interp(image, points):
    """Looks up the pixel values in an image at a given point using bilinear
    interpolation."""

    # initiate the interpolation values to 0s
    if image.ndim == 3:
      interpolation = np.zeros((points.shape[0], image.shape[2]))
    else:
      interpolation = np.zeros(points.shape[0])
    
    # calculate the valid points, which are the points staying inside the image
    valid_idx = ((points[:,0] <= image.shape[0]-1) \
                 * (points[:,0] >= 0) \
                 * (points[:,1] <= image.shape[1]-1) \
                 * (points[:,1] >= 0)).nonzero()
    valid_points = points[valid_idx]

    # Compute the four integer corner coordinates (top-left/right,
    # bottom-left/right) for interpolation, 
    bottom_left = np.floor(valid_points).astype(np.int32)
    bottom_right = bottom_left + np.asarray([[1, 0]])
    top_left = bottom_left + np.asarray([[0, 1]])
    top_right = bottom_left + np.asarray([[1, 1]])  #np.ceil(valid_points).astype(np.int32)

    # the fractional part of a, b the coordinates.
    ab = valid_points - bottom_left
    if image.ndim == 3: # case of 3 channel
      ab = np.expand_dims(ab, axis=-1)
    
    # interpolate between the four corner points
    # using the formulation in page 46 of CV_2020Spr_W3.pdf
    valid_interp = (1 - ab[:, 0]) * (1 - ab[:, 1]) * image[bottom_left[:, 0], bottom_left[:, 1]]  \
                    + ab[:, 0] * (1 - ab[:, 1]) * image[bottom_right[:, 0], bottom_right[:, 1]] \
                    + ab[:, 0] * ab[:, 1] * image[top_right[:, 0], top_right[:, 1]] \
                    + (1 - ab[:, 0]) * ab[:, 1] * image[top_left[:, 0], top_left[:, 1]]

    # only valid points get the interpolation values
    # invalid points (outside the image) got values of 0
    interpolation[valid_idx] = valid_interp
    return interpolation
```

```python
# normally blend with a mask
def blend_with_mask(source, target, mask):
    """Blends the source image with the target image according to the mask.
    Pixels with value "1" are source pixels, "0" are target pixels, and
    intermediate values are interpolated linearly between the two."""

    output = source * mask + target * (1 - mask)

    return output
    
```

For compositing a planar region in the source image to a region in the target image: \
+) Compute the bouding box src_BB of the masked region in source \
+) project src_BB to the target, then compute the bounding box tar_BB of the projection \
+) compute src_in_tar: warped image of the corresponding region of source into tar_BB \
+) compute mask_in_tar: warped image of the corresponding region of mask into tar_BB \
+) then blending src_in_tar with the region of target defined by tar_BB using mask_in_tar \

```python
def composite(source, target, source_pts, target_pts, mask, blender=blend_with_mask):
    """Composites a masked planar region of the source image onto a
    corresponding planar region of the target image via homography warping."""

    # make a copy of target, so that modifying target 
    # inside the function does not affect the original target
    target = target.copy()

    # make sure source and target are 3-D
    if source.ndim < 3:
      source = source[:,:,np.newaxis]
    if target.ndim < 3:
      target = target[:,:,np.newaxis]

    # Compute the homography to warp points from the target to the source coordinate frame.
    H_tar_src = compute_H(target_pts, source_pts)
    # the inverse homography from source to target
    H_src_tar = np.linalg.inv(H_tar_src)

    # compute points in the masked planar region of the source: the region where mask > 0
    if mask.ndim == 3:
      planar_x, planar_y = np.nonzero(np.sum(mask, axis=2))
    else:
      planar_x, planar_y = np.nonzero(mask)

    # calculate the bouding box (rectangular) containing the planar region of the source
    planar_minx, planar_maxx = np.min(planar_x), np.max(planar_x)
    planar_miny, planar_maxy = np.min(planar_y), np.max(planar_y)

    src_BB = np.asarray([[planar_minx, planar_miny], [planar_minx, planar_maxy], [planar_maxx, planar_miny], [planar_maxx, planar_maxy]])
    
    # project the bounding box containing the planar region of the source to the target
    target_PTS = project_Homo(H_src_tar, src_BB).astype(np.int32)

    # compute the bounding box in the target containing the above projection (target_PTS)
    tar_minx, tar_maxx = np.min(target_PTS[:,0]), np.max(target_PTS[:,0])
    tar_miny, tar_maxy = np.min(target_PTS[:,1]), np.max(target_PTS[:,1])
    
    # compute the bouding box shape in the target
    tar_BB_shape = (tar_maxx - tar_minx+1, tar_maxy - tar_miny + 1, target.shape[2])
    
    # the inverse projection of the bounding box in the target into the source
    tar_x, tar_y = np.meshgrid(range(tar_minx, tar_maxx+1), range(tar_miny, tar_maxy+1), indexing='ij')
    tar_x, tar_y = tar_x.flatten(), tar_y.flatten()
    tar_pts = np.ones((tar_x.shape[0], 2))
    tar_pts[:,0] = tar_x
    tar_pts[:,1] = tar_y
    src_pts = project_Homo(H_tar_src, tar_pts)

    # looking for pixel values in source and mask
    pixel_val = bilinear_interp(source, src_pts)
    src_in_tar = pixel_val.reshape(tar_BB_shape)

    mask_pixel = bilinear_interp(mask, src_pts)
    mask_in_tar = mask_pixel.reshape(tar_BB_shape[0], tar_BB_shape[1], mask.shape[2])
    
    # Blend the warped images with function blender (default by blend_with_mask)
    blended_tar = blender(src_in_tar, target[tar_minx:tar_maxx+1, tar_miny:tar_maxy+1], mask_in_tar)
    target[tar_minx:tar_maxx+1, tar_miny:tar_maxy+1] = blended_tar    

    return target

    
```

# Implement Poisson Image Blending

Note that: this implementation only considers the case where any pixel inside the mask has full 4 neighbor pixels. (I have not considered the case when some pixels have less than 4 neighbor pixels, actually I dont know in which scenario these type of pixels exist; due to the time constraint, this exception is left out.)

```python
"""
the mask here is a binary mask (actually, it can contain some gray value between 0 1 at the edge due to image warping and interpolation)
the mask we talking here is the one projected to fit in the region in the target image, not the original one with size of source image
Anyway, we only care about 0 or non-0 values, so there is no problems
"""

def boundary2zero(mask):
  """
  Find the location of the boundary pixels in mask, and set their values to 0
  In doing so, the corresponding boundary pixels in source image 
  can be seen as known pixels => can take values from the target image
  """

  # copy mask so that we dont modify mask outside the function
  mask = mask.copy()

  # pad 0 around mask
  if mask.ndim == 3:
    padded_mask = np.zeros((mask.shape[0]+2, mask.shape[1]+2, mask.shape[2]))
  else:
    padded_mask = np.zeros((mask.shape[0]+2, mask.shape[1]+2))
  
  padded_mask[1:-1,1:-1] = mask

  # the boundary pixels met when going left to right
  lr = np.where(mask - padded_mask[:-2, 1:-1] == 1)
  # right to left
  rl = np.where(mask - padded_mask[2:, 1:-1] == 1)
  # top to bottom
  tb = np.where(mask - padded_mask[1:-1, :-2] == 1)
  # bottom to top
  bt = np.where(mask - padded_mask[1:-1, 2:] == 1)

  # set those pixels to 0
  mask[lr] = 0
  mask[rl] = 0
  mask[tb] = 0
  mask[bt] = 0

  return mask

"""
# a test for boundary2zero
n = 10
m = 10
c = 1

mask = (np.random.rand(n, m, c) > 0.2).astype(np.uint8)
print(mask[:,:,0])

boundary2zero(mask)[:,:,0]
"""
```

**Building poisson matrix** using sparse matrix in scipy

```python
from scipy.sparse import lil_matrix
from scipy import sparse

def  build_poisson_matrix(n, m):
  """build poisson matrix for rectangular region of size nxm"""

  psn_mtx = (sparse.eye(m*n) * 4).tolil()
  idx1, idx2 = zip(*[(i, i+1) for i in range(n*m-1)])
  psn_mtx[idx1, idx2] = -1
  idx1, idx2 = zip(*[(i+1, i) for i in range(n*m-1)])
  psn_mtx[idx1, idx2] = -1
  idx1, idx2 = zip(*[(i, i+m) for i in range(0, n*m-m)])
  psn_mtx[idx1, idx2] = -1
  idx1, idx2 = zip(*[(i+m, i) for i in range(0, n*m-m)])
  psn_mtx[idx1, idx2] = -1

  idx1, idx2 = zip(*[(i*m, i*m-1) for i in range(0,n)])
  psn_mtx[idx1, idx2] = 0
  idx1, idx2 = zip(*[(i*m-1, i*m) for i in range(0,n)])
  psn_mtx[idx1, idx2] = 0
  
  return psn_mtx.tocsr()

# a test case
#%timeit build_poisson_matrix(500,400)
poisson_mat = build_poisson_matrix(2,2)
print(poisson_mat.toarray())
```

***Poisson Image Blending*** \
The idea of poisson image editting is to find new pixel values in the inner of the masked region of source image, so that the gradient remains the same while the absolute pixel values are consistent with the pixel values of the target image. \
I refer for a simple theoritical explanation (just the theory, not the Java code part) in this $\href{http://www.ctralie.com/Teaching/PoissonImageEditing/}{blog}$. 
I derive the following formulation by myself. \

For an image img of size $(n+1)\times(m+1)$, the gradient of the inner (rectangular region) can be computed by: \
$\nabla \text{inner} = A @ \text{inner_vec} + b$ \
Here, $A$ is the Poisson matrix of size $nm\times nm$, inner_vec is the $(nm, )$ vector of pixel values in the inner region, $b$ is a $(nm, )$ vector containing pixel values in the boundary (the pixel values arranged appropriately in $b$). \
We want, \
$\nabla \text{source_inner} = \nabla \text{poisson_inner}$ or, \

$A @ \text{source_inner_vec} + b_1 = A @ x + b_2$. Here, $b_1$ containing boundary pixels of source image, $b_2$ containing boundary pixels of target image, and x is the unknown poisson_inner_vec. \
Rewrite as, \

$A @ x = b$, for $b = b_0 + b_1 - b_2$ and $b_0 = A @ \text{source_inner_vec}$ \

By solving the above equation, we can find the pixel values of poisson_inner. \

When the ***source region is masked***, we can derive the above equation only for the inner of the masked region as follow: \
+) Derive the above equation for the whole inner of source \
+) Slice out the columns and rows corresponding to the known pixels (unmasked pixels + boundary pixels of the mask) \
+) The remaining is the needed equation for the unknown pixel values: $\text{A_unknown} @ \text{x_unknown} = \text{b_unknown}$. \




```python
from scipy.sparse.linalg import spsolve
from time import time

def poisson_blending(source, target, mask):
  """source, target and mask are of a same size"""

  t0 = time()
  print('In Poisson Blending: \n   source shape: %s - target shape: %s - mask shape: %s' \
        %(source.shape, target.shape, mask.shape))
  
  # initialize pois_source to equal target
  pois_source = np.copy(target) 
  # the inner region, modify later
  pois_inner_vec = pois_source[1:-1, 1:-1].reshape(-1, pois_source.shape[2])

  # extract source inner and find its shape
  src_inner = source[1:-1,1:-1]
  n, m = src_inner.shape[0], src_inner.shape[1]
  # vectors of pixel values in the inner of source
  src_inner_vec = src_inner.reshape(-1, src_inner.shape[2])

  # poisson matrix of size nxm
  A = build_poisson_matrix(n, m)

  # compute b0
  b0 = A @ src_inner_vec  
  # compute b1
  b1 = np.zeros_like(b0)
  b1[0:m] += -source[0, 1:-1]
  b1[(n-1)*m:n*m] += -source[-1, 1:-1]
  b1[0::m] += -source[1:-1, 0]
  b1[m-1::m] += -source[1:-1, -1]
  # compute b2
  b2 = np.zeros_like(b0)
  b2[0:m] += -target[0, 1:-1]
  b2[(n-1)*m:n*m] += -target[-1, 1:-1]
  b2[0::m] += -target[1:-1, 0]
  b2[m-1::m] += -target[1:-1, -1]
  # compute b
  b = b0 + b1 - b2
  
  # uncomment to check if grad_source == grad_poisson
  #x = spsolve(A, b)
  #grad_poisson = (A @ x).reshape(n, m, src_inner.shape[2]) + b2.reshape(n, m, src_inner.shape[2])
  #grad_source = (b0 + b1).reshape(n, m, -1)
  #print('grad_source == grad_poisson: ', np.allclose(grad_source, grad_poisson))
  
  # update mask so that the unmasked and boundary pixels -> 0
  # Hence, those corresponding pixels in pois_source become known pixels.
  mask = boundary2zero(mask)

  mask_inner = mask[1:-1, 1:-1] #just to fit size with src_inner
  mask_vec = mask_inner.reshape(-1) #assuming mask is 2D or 3D of shape (x,y,1)

  # known pixels 
  known_pixel = np.where(mask_vec == 0)[0] # idx of know pixel in pois_source
  # unknown pixels
  unknown_pixel = np.where(mask_vec > 0)[0]

  # slice out known columns and update b
  known_A = A[:, known_pixel] 
  known_x = pois_inner_vec[known_pixel]
  b -= known_A @ known_x

  # select the unknown columns of A
  unknown_A = A[:, unknown_pixel]
  # select the unknown rows of A and b
  unknown_b = b[unknown_pixel]
  unknown_A = unknown_A[unknown_pixel]
  # solve equation to find unknown_x
  unknown_x = spsolve(unknown_A, unknown_b)

  pois_inner_vec[unknown_pixel] = unknown_x.reshape(pois_inner_vec[unknown_pixel].shape)
  pois_source[1:-1, 1:-1] = pois_inner_vec.reshape(n, m, pois_source.shape[2])

  print('   blending in %.2f s' %(time() - t0))
  return pois_source


# simple test case
n1 = 5
m1 = 5
ch = 3

src = np.ones((n1, m1, ch))*2
src[1:-1,1:-1] = np.ones((n1-2,m1-2, ch)) * 7
src[2:-2,2:-2] = np.ones((n1-4,m1-4, ch)) * -3
print('source: \n', src[:,:,0])

tar = np.ones((n1,m1,ch))*3
print('target: \n', tar[:,:,0])
msk = np.ones((n1, m1, 1))
print('mask: \n', msk[:,:,0])

pois_source = poisson_blending(src, tar, msk)
print('pois_source:\n', pois_source[:,:,0])
```
